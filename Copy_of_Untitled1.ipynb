```python
def get_pay(num_hours):
    #pre-tax pay,based on receiving BDT 2000/hr
    pay_pretax=num_hours*2000
    # after-tax pay,based on being 12% tax bracket
    pay_aftertax=pay_pretax*(1-0.12)
    return pay_aftertax
```
